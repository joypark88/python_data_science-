```python
#Step 1. Import the necessary libraries
import pandas as pd
```

```python
#Step 2. Import the dataset from this address.
#Step 3. Assign it to a variable called drinks.

drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks
```

```python
#Step 4. Which continent drinks more beer on average?
beer_mean = drinks.beer_servings.mean()
beer_mean
drinks.loc[drinks.beer_servings > beer_mean]
```

```python
#Step 5. For each continent print the statistics for wine consumption
drinks.iloc[:,[3,5]].groupby('continent').sum()
```

```python
#Step 6. Print the mean alcohol consumption per continent for every column
continent_alcohol = drinks.iloc[:,[4,5]].groupby('continent').mean()
continent_alcohol.columns =  ['alcohol_mean']
drinks = pd.merge(drinks, continent_alcohol, left_on='continent', right_on='continent', how = 'left')
drinks

#another answer
drinks.groupby('continent').mean()
```

```python
#Step 7. Print the median alcohol consumption per continent for every column
continent_alcohol = drinks.iloc[:,[4,5]].groupby('continent').median()
continent_alcohol.columns =  ['alcohol_median']
drinks = pd.merge(drinks, continent_alcohol, left_on='continent', right_on='continent', how = 'left')
drinks

#drinks.groupby('continent').median()
```

```python
#Step 8. Print the mean, min and max values for spirit consumption.
#This time output a DataFrame
# drinks.alcohol_median.mean()
# drinks.alcohol_median.min()
# drinks.alcohol_median.max()

drinks['alcohol_median'].agg(['max','min','mean'])

#Answer
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```
